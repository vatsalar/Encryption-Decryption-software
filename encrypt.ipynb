```python
#Encryption for first video
from Crypto.Cipher import AES
import hashlib
import os
#from os import urandom


password = b'mypassword'
key = hashlib.sha256(password).digest()
mode = AES.MODE_CBC
IV = 'This is an IV456'

def pad_message(file):
    while len(file) % 16 != 0:
        file = file + b'0'
    return file

#object of aes
cipher = AES.new(key, mode, IV)

#file handling
with open('test.mp4','rb') as f:
    orig_file = f.read()





padded_file = pad_message(orig_file)


encrypted_message = cipher.encrypt(padded_file)

#file handling
with open('encrypted_file2', 'wb')as e:
    e.write(encrypted_message)
    

 """"CODE FOR SECOND FILE"""   
#Encryption for second video
 from Crypto.Cipher import AES
import hashlib
import os
#from os import urandom


password = b'mypassword'
key = hashlib.sha256(password).digest()
mode = AES.MODE_CBC
IV = 'This is an IV456'

def pad_message(file):
    while len(file) % 16 != 0:
        file = file + b'0'
    return file


cipher = AES.new(key, mode, IV)

with open('testing.mp4','rb') as f:
    orig_file = f.read()





padded_file = pad_message(orig_file)


encrypted_message = cipher.encrypt(padded_file)

with open('encrypted_video_file2', 'wb')as e:
    e.write(encrypted_message)   
```
